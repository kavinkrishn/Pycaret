## Missing Values 

```python
from pycaret.datasets import get_data
```

```python
hepatitis = get_data("hepatitis")
```

```python
setup(data=hepatitis,target='Class')   
```

```python
pokemon = get_data("pokemon")
```

```python
setup(data=pokemon,target='Legendary',ignore_features=['#','Name'])  # Ignore the two columns
```

## Change the number to Categorical

```python
clf = setup(data=hepatitis,target='Class',categorical_features=['AGE'])   # to change countious to categorical values in one hot encoding
```

## Ordinal Encodeing

```python
## assign the categorical values to rank values
```

```python
employee = get_data("employee")
```

```python
employee['salary'].value_countss()
```

```python
clf = setup(data=employee,target='left',ordinal_features={'salary':['low','medium','high']})
```

```python
clf
```
